```python
import pandas as pd
import random

# Lista de productos
productos = [
    "Micas protectoras", "Fundas variadas", "Cargadores USB", "Power banks",
    "Auriculares cable", "Auriculares Bluetooth", "Adaptadores audífonos",
    "Soportes teléfono", "PopSockets y anillos", "Tarjetas microSD",
    "Protectores lente", "Lentes smartphone", "Stylus táctiles",
    "Baterías reemplazo", "Kits reparación", "Cable USB C/Lightning",
    "Adaptadores OTG", "Selfie sticks", "Luces LED selfie",
    "Protectores teclado", "Skins y stickers", "Bolsas impermeables",
    "Cintas y correas", "Parlantes Bluetooth", "Adaptadores inalámbricos",
    "Adaptadores TV", "Tarjeteros teléfono", "Cámaras smartphone",
    "Mini trípodes", "Cables retráctiles", "Alfombrillas inalámbricas",
    "Protectores puertos", "Kit limpieza celular", "Abrazaderas y clips",
    "Bolsos y mochilas", "Guantes táctiles", "Cinturones deportivos",
    "Cables carga rápida", "Soportes magnéticos", "Anillos soporte teléfono"
]

# Generar cantidades de productos vendidos aleatoriamente hasta sumar 1500
total_productos = 1500
cantidades_vendidas = []
while total_productos > 0:
    cantidad = random.randint(1, min(total_productos, 50))  # Se venden hasta 50 productos a la vez
    cantidades_vendidas.append(cantidad)
    total_productos -= cantidad

# Generar días de la semana de venta
dias_semana = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado"]
dias_vendidos = [random.choice(dias_semana) for _ in range(len(cantidades_vendidas))]

# Generar precios de los productos en soles
precios = [
    20,  # Micas protectoras
    25,  # Fundas variadas
    15,  # Cargadores USB
    40,  # Power banks
    10,  # Auriculares cable
    80,  # Auriculares Bluetooth
    5,   # Adaptadores audífonos
    12,  # Soportes teléfono
    8,   # PopSockets y anillos
    30,  # Tarjetas microSD
    18,  # Protectores lente
    50,  # Lentes smartphone
    6,   # Stylus táctiles
    35,  # Baterías reemplazo
    15,  # Kits reparación
    8,   # Cable USB C/Lightning
    5,   # Adaptadores OTG
    12,  # Selfie sticks
    15,  # Luces LED selfie
    7,   # Protectores teclado
    3,   # Skins y stickers
    20,  # Bolsas impermeables
    10,  # Cintas y correas
    60,  # Parlantes Bluetooth
    25,  # Adaptadores inalámbricos
    40,  # Adaptadores TV
    8,   # Tarjeteros teléfono
    70,  # Cámaras smartphone
    20,  # Mini trípodes
    5,   # Cables retráctiles
    15,  # Alfombrillas inalámbricas
    4,   # Protectores puertos
    10,  # Kit limpieza celular
    3,   # Abrazaderas y clips
    30,  # Bolsos y mochilas
    8,   # Guantes táctiles
    15,  # Cinturones deportivos
    10,  # Cables carga rápida
    18,  # Soportes magnéticos
    10,  # Anillos soporte teléfono
]

# Asegurarnos de que ambas listas tengan la misma longitud usando zip
data = list(zip(productos, cantidades_vendidas, dias_vendidos, precios))

# Crear un DataFrame con los datos
df = pd.DataFrame(data, columns=["Producto", "Cantidad vendida", "Día de la semana de venta", "Precio"])

# Guardar el DataFrame en un archivo Excel
nombre_archivo = "ventas_productos_celulares.xlsx"
df.to_excel(nombre_archivo, index=False)

print("Archivo Excel creado con éxito: ", nombre_archivo)
```
